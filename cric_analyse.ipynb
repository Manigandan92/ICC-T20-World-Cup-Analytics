```python
total_runs_series = deliveries['total_runs'].sum()
total_wickets_series = deliveries['wicket_type'].notna().sum()

print("===== SERIES SUMMARY =====")
print("Total Runs in Series:", total_runs_series)
print("Total Wickets in Series:", total_wickets_series)
```

```python
batsman_runs = deliveries.groupby('striker')['runs_off_bat'].sum()
batsman_runs = batsman_runs.sort_values(ascending=False)
```

```python
dismissals = deliveries[deliveries['wicket_type'].notna()]
dismissals = dismissals.groupby('player_dismissed')['wicket_type'].count()
```

```python
# batting average
batting_average = batsman_runs / dismissals
batting_average = batting_average.replace([np.inf], np.nan)
batting_average = batting_average.dropna().sort_values(ascending=False)
```

```python
#Bowling Analysis
bowler_wickets = deliveries[deliveries['wicket_type'].notna()]
bowler_wickets = bowler_wickets.groupby('bowler')['wicket_type'].count()
bowler_wickets = bowler_wickets.sort_values(ascending=False)

# Economy Rate
bowler_runs = deliveries.groupby('bowler')['total_runs'].sum()
bowler_balls = deliveries.groupby('bowler').size()

economy_rate = (bowler_runs / bowler_balls) * 6
economy_rate = economy_rate.sort_values()

print("\n===== TOP 5 BOWLERS (Wickets) =====")
print(bowler_wickets.head(5))

print("\n===== BEST ECONOMY RATE =====")
print(economy_rate.head(5))
```

```python
import pandas as pd
import numpy as np

# Load data
deliveries = pd.read_csv('deliveries.csv')
matches = pd.read_csv('matches.csv')

# Create total_runs column
deliveries['total_runs'] = deliveries['runs_off_bat'] + deliveries['extras']

# Get unique match IDs
match_ids = deliveries['match_id'].unique()

for match in match_ids:
    
    print("\n" + "="*60)
    print(f"üèè MATCH ID: {match}")
    print("="*60)
    
    match_data = deliveries[deliveries['match_id'] == match]
    
    # --------------------------
    # Team Total Score
    # --------------------------
    team_total = match_data.groupby('batting_team')['total_runs'].sum()
    print("\nüìä Team Totals:")
    print(team_total)
    
    # --------------------------
    # Individual Batting Score
    # --------------------------
    batting_score = match_data.groupby('striker')['runs_off_bat'].sum()
    batting_score = batting_score.sort_values(ascending=False)
    
    print("\nüèè Batting Scorecard:")
    print(batting_score)
    
    # --------------------------
    # Individual Bowling Figures
    # --------------------------
    wickets = match_data[match_data['wicket_type'].notna()]
    bowler_wickets = wickets.groupby('bowler')['wicket_type'].count()
    
    overs = match_data.groupby('bowler').size() / 6
    
    bowling_scorecard = pd.concat([bowler_wickets, overs], axis=1)
    bowling_scorecard.columns = ['Wickets', 'Overs']
    bowling_scorecard = bowling_scorecard.fillna(0)
    bowling_scorecard = bowling_scorecard.sort_values(by='Wickets', ascending=False)
    
    print("\nüéØ Bowling Figures:")
    print(bowling_scorecard)
    
    print("\n" + "="*60)
```

```python
match_totals = deliveries.groupby('match_id')['total_runs'].sum().reset_index()

X = match_totals[['match_id']]
y = match_totals['total_runs']

model = LinearRegression()
model.fit(X, y)

match_totals['predicted_runs'] = model.predict(X)

print("\n===== ML MODEL SAMPLE OUTPUT =====")
print(match_totals.head())
```
